
# Lineárne funkcie

Lineárna funkcia je určená rovnicou
$$y = ax+b,$$
kde $a \ne 0,b$ sú reálne čísla.
Definičným oborom je množina reálnych čísel.
Grafom je priamka so smernicou $a$ prechádzajúca bodom $[0, b]$.

* [Wikipédia](https://en.wikipedia.org/wiki/Linear_function_(calculus))

V nasledujúcich príkladoch budeme kresliť grafy a vyšetrovať priebeh lineárnych funkcií.

## Úvodný príklad

Nakreslenie grafu lineárnej funkcie danej rovnicou
$$y = x.$$

```python
#####
##### nakreslenie grafu funkcie
#####

#### vstupné údaje
def f(X): return X # ufunc verzia funkcie
X = np.linspace(-3, 3, 6*10+1) # výber hodnôt nezávislej premennej pre zaujímavú časť grafu
Y = f(X) # odpovedajúce hodnoty závislej premennej

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(6,6) # veľkosť obrázka (východzia hodnota je 6x4)

### diagram

## inicializácia diagramu: vytvorí sa pravoúhla súradnicová sústava
init_subplot(ax)

## pomenovanie diagramu
ax.set_title(r"Graf funkcie $y = x$")

## nastavenie rovnakej mierky pre obe osi
ax.set_aspect('equal')

## pravoúhla sieť
#ax.grid()

## graf funkcie
ax.plot(X, Y)

### archivácia obrázka
#plt.savefig("<meno súboru>.png")

### samotné zobrazenie
plt.show()
plt.close('all')
```

## Trochu zložitejší príklad

Nakreslenie grafu a vyšetrenie priebehu lineárnej funkcie danej rovnicou
$$y = 2x+3.$$
Tu nás zaujíma tiež táto informácia:
* Nulové body funkcie.

```python
#####
##### nakreslenie grafu a vyšetrenie priebehu funkcie
#####

#### vstupné údaje
def f(X): return 2 * X + 3 # ufunc verzia funkcie
X = np.linspace(-3, 1, 4*10+1) # výber hodnôt nezávislej premennej pre zaujímavú časť grafu
Y = f(X) # odpovedajúce hodnoty závislej premennej

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(4,8) # veľkosť obrázka (východzia hodnota je 6x4)

### diagram

## inicializácia diagramu: vytvorí sa pravoúhla súradnicová sústava
init_subplot(ax)

## pomenovanie diagramu
ax.set_title(r"Graf a priebeh funkcie $y = 2x+3$")

## nastavenie rovnakej mierky pre obe osi
ax.set_aspect('equal')

## pravoúhla sieť
ax.grid()

## graf funkcie
ax.plot(X, Y)

## nulové body funkcie
ax.plot(-3/2, 0, 'o', label="nulový bod")
ax.annotate(r"$\left[-\frac{3}{2},0\right]$", xy=(-3/2, 0), va='bottom', ha='right')

## legenda
ax.legend()

### archivácia obrázka
#plt.savefig("<meno súboru>.png")

### samotné zobrazenie
plt.show()
plt.close('all')
```

## Úloha

Nakreslite grafy a vyšetrite priebeh týchto funkcií
$$y = 2x-3$$
$$y = 3x+1$$
$$y = x+4$$
$$y = 4x-2$$
$$y = 7-x$$
$$y = 2x+3$$
$$y = 2-3x$$
$$y = 5x-1$$

```python
def vykresli_a_vysetri(X, Y, predpis, nulovy_bod=(0, 0), farba='red'):
    
    nb_x, nb_y = nulovy_bod
    
    #### obrázok s jedným diagramom
    fig, ax = plt.subplots()
    fig.set_size_inches(6,6)
    
    ### diagram
    init_subplot(ax)
    ax.set_title(r"Graf a priebeh funkcie " + predpis)
    ax.set_aspect('equal')
    ax.grid()
    
    ## graf funkcie
    ax.plot(X, Y, c='r')
    
    ## nulové body funkcie
    ax.plot(nb_x, nb_y, 'o', c='lightgreen', markersize=10, label="nulový bod")
    ax.annotate(f"$[{nb_x},{nb_y}]$", xy=(nb_x, nb_y))
    
    ## legenda
    ax.legend()
    
    ### samotné zobrazenie
    plt.show()
    plt.close('all')
    
```

## Moje riešenie:

```python
def f1(X): return 2*X - 3
def f2(X): return 3*X - 1
def f3(X): return X + 4
def f4(X): return 4*X - 2
def f5(X): return 7 - X
def f6(X): return 2*X + 3
def f7(X): return 2 - 3*X
def f8(X): return 5*X - 1

farby = [f'#{random.randrange(256**3):06x}' for i in range(8)]
predpisy = ['y = 2x-3', 'y = 3x+1', 'y = x+4', 'y = 4x-2', 'y = 7-x', 'y = 2x+3', 'y = 2-3x', 'y = 5x-1']
nulove_body = [(3/2, 0), (-1/3, 0), (-4, 0), (1/2, 0), (7, 0), (-3/2, 0), (2/3, 0), (1/5, 0)]

X = np.linspace(-10, 10, 10+1)

for f, color, predpis, nulovy_bod in zip([f1, f2, f3, f4, f5, f6, f7, f8], farby, predpisy, nulove_body):
    Y = f(X)
    vykresli_a_vysetri(X, Y, predpis=predpis, nulovy_bod=nulovy_bod, farba=color)
```

## Úloha

Nájdite lineárnu funkciu, ktorej graf je priamka prechádzajúca bodom $[2,2]$ rovnobežná s  priamkou $x+2y-3 = 0$. Overte graficky, že vaše riešenie je správne.  

## Moje riešenie

Funkcia vychádzajúca z predpisu priamky $x+2y-3 = 0$ je $y=\frac{3-x}{2}$. Jej prienik s osou $y$, označme $b_1 = \frac{3}{2}$.

V hľadanej funkcii je to zjavne dvojnásobok, teda $b_2 = \frac{3}{2} \times 2 = 3$. Teraz podobne ako v predošlej úlohe, už máme 2 body - (2, 2) a (0, 3). Vypočítajme smernicu $a_2 = \frac{y2-y1}{x2-x1} = -\frac{1}{2}$.

Hľadaná funkcia tak má predpis $y=-\frac{1}{2}x + 3$.

Overme:

```python
###
### šablóna riešenia
###

## zadanie
fig, ax = plt.subplots()
fig.set_size_inches(8,6)
init_subplot(ax)
ax.set_aspect('equal')
ax.grid()

ax.plot(2, 2, 'o', markersize=10, c='lightgreen')
X = np.linspace(-1, 7, 8*10+1)
ax.plot(X, (3 - X) / 2, label=r"priamka $x+2y-3 = 0$")

## riešenie
def f(X): return (-1/2) * X + 3
Y = f(X)
ax.plot(X, Y, label="riešenie y = -1/2x + 3", c='red')

## dokončenie
ax.legend()
plt.show()
plt.close('all')
```

## Úloha (3 body)

Nájdite lineárnu funkciu, ktorej graf je priamka prechádzajúca bodom $[1,2]$ kolmá na priamku $3x-4y-2 = 0$. Overte graficky, že vaše riešenie je správne.  

## Moje riešenie

Funkcia vychádzajúca z predpisu priamky $x+2y-3 = 0$ je $y=\frac{3}{4}x - \frac{1}{2}$. Smernica tejto priamky je $a_1=\frac{3}{4}$. Aby bola hľadaná priamka kolmá na našu priamku, ich smernice musia byť vo vzťahu $a_1 \cdot a_2 = -1$. Dosadením dostaneme $a_2 = -\frac{4}{3}$. Teraz už len využitím zadaného bodu $(1, 2)$ vypočítajme prienik s osou $y$, $b_2=2 + \frac{4}{3} = \frac{10}{3}$.

Hľadaná funkcia tak má predpis $y=-\frac{4}{3}x + \frac{10}{3}$.

Overme:

```python
###
### šablóna riešenia
###

## zadanie
fig, ax = plt.subplots()
fig.set_size_inches(4,6)
init_subplot(ax)
ax.set_aspect('equal')
ax.grid()

ax.plot(1, 2, 'o', markersize=10, c='lightgreen')
X = np.linspace(-1, 3, 4*10+1)
ax.plot(X, 3/4 * X - 1/2, label=r"priamka $3x-4y-2 = 0$")

## riešenie
def f(X): return (-4/3) * X + 10/3
Y = f(X)
ax.plot(X, Y, label="riešenie", c='red')

## dokončenie
ax.legend()
plt.show()
plt.close('all')
```

## Úloha (3 body)

Nájdite lineárnu funkciu, ktorej graf je dotyčnica kružnice
$x^2+y^2 = 1$ v bode $[\cos \frac{\pi}{4},\sin \frac{\pi}{4}]$. Overte graficky, že vaše riešenie je správne.  

## Moje riešenie

Z predpisu kružnice $x^2+y^2 = 1$ jasne vyplýva, že jej stred je v bode $(0, 0)$. Vďaka tomu máme dva body $(0,0)$ a $(\cos \frac{\pi}{4},\sin \frac{\pi}{4})$. Smernica priamky prechádzajucej cez tieto dva body je $1$. Hľadaná priamka je kolmá na túto priamku. Už vieme, že $a_1 \cdot a_2 = -1$, ak majú byť navzájom kolmé. Teda smernica hľadanej priamky je $a_2=-1$. 

Môžme dosadiť do $y=ax + b$, teda $\sin \frac{\pi}{4} = -\cos \frac{\pi}{4} + b_2$, z čoho máme $b_2 = \sqrt{2}$.

Hľadaná funkcia tak má predpis $y=-x + \sqrt{2}$.

Overme:

```python
###
### šablóna riešenia
###

## zadanie
fig, ax = plt.subplots()
fig.set_size_inches(5,6)
init_subplot(ax)
ax.set_aspect('equal')
#ax.grid()

T = np.linspace(0, 2 * np.pi, 100)
ax.plot(np.cos(T), np.sin(T), label="kružnica")
ax.plot(np.cos(np.pi/4), np.sin(np.pi/4), 'o', label=r"bod $\left[\cos\,\frac{\pi}{4},\sin\,\frac{\pi}{4}\right]$",
       markersize=10, c='lightgreen')

## riešenie
def f(X): return -X + np.sqrt(2)
X = np.linspace(-2, 3, 5*10+1)
Y = f(X)
ax.plot(X, Y, label="riešenie", c='red')

## dokončenie
ax.legend()
plt.show()
plt.close('all')
```

## Príklad

Nakreslenie grafov lineárnych homogénnych funkcií
$$y = ax$$
pre $a = -2, -1, -\frac{1}{2}, \frac{1}{2}, 1, 2$ do jedného obrázka.

Dokumentácia:

* [matplotlib.axes.Axes.set_xticks](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_xticks.html)
* [matplotlib.axes.Axes.set_yticks](https://matplotlib.org/api/_as_gen/matplotlib.axes.Axes.set_yticks.html)

```python
#####
##### grafy parametrického systému funkcií
#####

#### vstupné údaje
def f(X, a): return a * X
X = np.linspace(-3, 3, 6*10+1)

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(6,12)

### diagram
init_subplot(ax) 
ax.set_title(r"Grafy funkcií $y = ax$")
ax.set_aspect('equal')
ax.grid()
ax.set_xticks(range(-3,3+1)) # kótovanie x-ovej osi
ax.set_yticks(range(-6,6+1)) # kótovanie y-ovej osi

## grafy funkcií
for a in [-2, -1, -1/2, 1/2, 1, 2]:
    ax.plot(X, f(X, a), label=r"$a = {}$".format(a))

## legenda
ax.legend()

### archivácia obrázka
#plt.savefig("<meno súboru>.png")

### samotné zobrazenie
plt.show()
plt.close('all')
```

## Pokračovanie predchadzajúceho príkladu

To isté ako v predchadzajúcom príklade, len vykreslenie sa deje pomocou interaktívnych prvkov knižnice `ipywidgets`.

```python
#####
##### grafy parametrického systému funkcií (interaktívna verzia)
#####

#### vstupné údaje
def f(X, a): return a * X
X = np.linspace(-3, 3, 6*10+1)

#### obrázok s grafom funkcie y = ax
def plot_graph(a):
    fig, ax = plt.subplots()
    fig.set_size_inches(6,12)
    init_subplot(ax)
    ax.set_title(r"Graf funkcie $y = {0}{{x}}$".format(a))
    ax.grid()
    ax.set_ylim(-7, 7) # ohraničenie hodnôt pre osu y
    ax.set_xticks(range(-3,3+1)) # kótovanie x-ovej osi
    ax.set_yticks(range(-6,6+1)) # kótovanie y-ovej osi
    ax.plot(X, f(X, a))
    plt.show()
    plt.close('all')

#### interaktívne vykreslenie obrázka
import ipywidgets as widgets
widgets.interact(plot_graph, 
                 a=widgets.SelectionSlider(options=[-2, -1, -1/2, 1/2, 1, 2], value=1)
                 );
```

## Interaktívny príklad

Nakreslenie grafu lineárnej funkcie danej rovnicou vo vrcholovom tvare
$$y = a(x-h)+k$$
pre vybrané hodnoty parametrov $a,h,k$.
Grafom je priamka so smernicou $a$ prechádzajúca bodom $[h,k]$.

```python
#####
##### grafy parametrického systému funkcií (interaktívna verzia)
#####

#### vstupné údaje
def f(X, a, h, k): return a * (X - h) + k
X = np.linspace(-3, 3, 6*10+1)

#### obrázok s grafom funkcie y = a(x-h)+k
def plot_graph(a, h, k):
    fig, ax = plt.subplots()
    fig.set_size_inches(6,19)
    init_subplot(ax)
    ax.set_title(r"Graf funkcie $y = {0}({{x}}-{1})+{2}$".format(a, h, k))
    ax.set_aspect('equal')
    ax.grid()
    ax.set_ylim(-9.5, 9.5) # ohraničenie hodnôt pre osu y
    ax.set_yticks(range(-9,9+1)) # kótovanie y-ovej osi
    ax.plot(X, f(X, a, h, k))
    plt.show()
    plt.close('all')

#### interaktívne vykreslenie obrázka
import ipywidgets as widgets
widgets.interact(plot_graph,
                 a=widgets.SelectionSlider(options=[-2, -1, -1/2, 1/2, 1, 2], value=1),
                 h=widgets.IntSlider(min=-1, max=1, step=1, value=0),
                 k=widgets.IntSlider(min=-1, max=1, step=1, value=0)
                );
```

## Interaktívny príklad

Nakreslenie grafu lineárnej funkcie danej rovnicou v štandardnom tvare
$$y = ax+b$$
pre vybrané hodnoty parametrov $a,b$.

```python
#####
##### grafy parametrického systému funkcií (interaktívna verzia)
#####

#### vstupné údaje
def f(X, a, b): return a * X + b
X = np.linspace(-3, 3, 6*10+1)

#### obrázok s grafom funkcie y = ax+b
def plot_graph(a, b):
    fig, ax = plt.subplots()
    fig.set_size_inches(4,12)
    init_subplot(ax)
    ax.set_title(r"Graf funkcie $y = {0}{{x}}+{1}$".format(a, b))
    ax.set_aspect('equal')
    ax.grid()
    ax.set_ylim(-8.5, 8.5) # ohraničenie hodnôt pre osu y
    ax.set_yticks(range(-8,8+1)) # kótovanie y-ovej osi
    ax.plot(X, f(X, a, b))
    plt.show()
    plt.close('all')

#### interaktívne vykreslenie obrázka
import ipywidgets as widgets
widgets.interact(plot_graph, 
                 a=widgets.SelectionSlider(options=[-2, -1, -1/2, 1/2, 1, 2], value=1),
                 b=widgets.IntSlider(min=-2, max=2, step=1, value=0)
                );
```
