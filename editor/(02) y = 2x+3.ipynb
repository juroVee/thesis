```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.lines as ml
from maux import *
from jedit import editor

#### vstupné údaje
def f(X): return 2 * X + 3 # ufunc verzia funkcie
X = np.linspace(-3, 1, 4*10+1) # výber hodnôt nezávislej premennej pre zaujímavú časť grafu
Y = f(X) # odpovedajúce hodnoty závislej premennej

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(4,8) # veľkosť obrázka (východzia hodnota je 6x4)

### diagram

## inicializácia diagramu: vytvorí sa pravoúhla súradnicová sústava
init_subplot(ax)

## pomenovanie diagramu
ax.set_title(r"Graf a priebeh funkcie $y = 2x+3$")

## nastavenie rovnakej mierky pre obe osi
ax.set_aspect('equal')

## pravoúhla sieť
ax.grid()

## graf funkcie
ax.plot(X, Y)

## nulové body funkcie
ax.plot(-3/2, 0, 'o', label="nulový bod")
ax.annotate(r"$\left[-\frac{3}{2},0\right]$", xy=(-3/2, 0), va='bottom', ha='right')

## legenda
ax.legend()

### archivácia obrázka
#plt.savefig("<meno súboru>.png")

### samotné zobrazenie
plt.show()
#plt.close('all')
```

```python
editor(figure=fig, axes=ax, function=f, intervals=[X])
```
