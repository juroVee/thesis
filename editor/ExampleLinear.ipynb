```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.lines as ml
from maux import *
from jedit import editor

#### vstupnÃ© Ãºdaje
def f(X): return 2 * X + 3 # ufunc verzia funkcie
X = np.linspace(-3, 1, 4*10+1) # vÃ½ber hodnÃ´t nezÃ¡vislej premennej pre zaujÃ­mavÃº ÄasÅ¥ grafu
Y = f(X) # odpovedajÃºce hodnoty zÃ¡vislej premennej

#### obrÃ¡zok s jednÃ½m diagramom
fig, ax = plt.subplots()
fig.set_size_inches(4,8) # veÄ¾kosÅ¥ obrÃ¡zka (vÃ½chodzia hodnota je 6x4)

### diagram

## inicializÃ¡cia diagramu: vytvorÃ­ sa pravoÃºhla sÃºradnicovÃ¡ sÃºstava
init_subplot(ax)

## pomenovanie diagramu
ax.set_title(r"Graf a priebeh funkcie $y = 2x+3$")

## nastavenie rovnakej mierky pre obe osi
ax.set_aspect('equal')

## pravoÃºhla sieÅ¥
ax.grid()

## graf funkcie
ax.plot(X, Y)

## nulovÃ© body funkcie
ax.plot(-3/2, 0, 'o', label="nulovÃ½ bod")
ax.annotate(r"$\left[-\frac{3}{2},0\right]$", xy=(-3/2, 0), va='bottom', ha='right')

## legenda
ax.legend()

### archivÃ¡cia obrÃ¡zka
#plt.savefig("<meno sÃºboru>.png")

### samotnÃ© zobrazenie
plt.show()
#plt.close('all')
```
