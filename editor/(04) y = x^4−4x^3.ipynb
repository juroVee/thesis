```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from maux import *
from jedit import editor

#####
##### nakreslenie grafu funkcie
#####

#### vstupné údaje
def f(X): return X ** 4 - 4 * X ** 3 # ufunc verzia funkcie
X = np.linspace(-0.5, 0.5, 29*100+1) # výber hodnôt nezávislej premennej pre zaujímavú časť grafu
Y = f(X) # odpovedajúce hodnoty závislej premennej

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(7, 9) # veľkosť obrázka (východzia hodnota je 6x4)

### diagram
init_subplot(ax) # inicializácia diagramu: vytvorí sa pravoúhla súradnicová sústava
ax.set_title(r"Graf funkcie $y = x^4-4x^3$") # pomenovanie diagramu
ax.set_aspect('equal') # nastavenie rovnakej mierky pre obe osi
ax.grid() # pravoúhla sieť
ax.get_xticks()

## graf funkcie
ax.plot(X, Y)
ax.plot(X, Y)

### archivácia obrázka
#fig.savefig("<meno súboru>.png")

### samotné zobrazenie
fig.show()

def d(X): return 4 * (X - 3) * X ** 2
def d2(X): return np.power(X, 3)
```

```python
editor(figure=fig, axes=ax, function=f, intervals=[X], primes=[d])
```
