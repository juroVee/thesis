```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from maux import *
from jedit import editor

#####
##### nakreslenie grafu funkcie
#####

#### vstupnÃ© Ãºdaje
def f(X): return 2 * X ** 4 - 7 * X ** 3 + 5 * X ** 2 + 3 * X - 2 # ufunc verzia funkcie
X = np.linspace(-0.70, 2.20, 29*100+1) # vÃ½ber hodnÃ´t nezÃ¡vislej premennej pre zaujÃ­mavÃº ÄasÅ¥ grafu
Y = f(X) # odpovedajÃºce hodnoty zÃ¡vislej premennej

#### obrÃ¡zok s jednÃ½m diagramom
fig, ax = plt.subplots()
fig.set_size_inches(7, 9) # veÄ¾kosÅ¥ obrÃ¡zka (vÃ½chodzia hodnota je 6x4)

### diagram
init_subplot(ax) # inicializÃ¡cia diagramu: vytvorÃ­ sa pravoÃºhla sÃºradnicovÃ¡ sÃºstava
ax.set_title(r"Graf funkcie $y = 2x^4-7x^3+5x^2+3x-2$") # pomenovanie diagramu
ax.set_aspect('equal') # nastavenie rovnakej mierky pre obe osi
ax.grid() # pravoÃºhla sieÅ¥
ax.get_xticks()

## graf funkcie
ax.plot(X, Y)

### archivÃ¡cia obrÃ¡zka
#fig.savefig("<meno sÃºboru>.png")

### samotnÃ© zobrazenie
fig.show()
```
