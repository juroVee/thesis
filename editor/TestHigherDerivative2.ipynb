```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from maux import *
from jedit import editor

#### vstupnÃ© Ãºdaje
def f(X): return np.power(X, 6) + 5
X = np.linspace(-2, 2, 100+1)
Y = f(X)

fig, ax = plt.subplots()
fig.set_size_inches(8, 6)

init_subplot(ax)
ax.set_title(r"Graf funkcie $y = x^6 + 5$")
ax.grid()

ax.plot(X, Y)

fig.show()
```
