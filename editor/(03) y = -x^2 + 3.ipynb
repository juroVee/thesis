```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from maux import *
from jedit import editor
from scipy.optimize import newton

#####
##### nakreslenie grafu funkcie
#####

#### vstupné údaje
def f(X): return -np.power(X, 2) + 3 # ufunc verzia funkcie
X = np.linspace(-5, 5, 1000+1) # výber hodnôt nezávislej premennej pre zaujímavú časť grafu
Y = f(X) # odpovedajúce hodnoty závislej premennej

#### obrázok s jedným diagramom
fig, ax = plt.subplots()
fig.set_size_inches(7, 9) # veľkosť obrázka (východzia hodnota je 6x4)

### diagram
init_subplot(ax) # inicializácia diagramu: vytvorí sa pravoúhla súradnicová sústava
ax.set_title(r"Graf funkcie $y = -x^2 + 3$") # pomenovanie diagramu
ax.set_aspect('auto') # nastavenie rovnakej mierky pre obe osi
ax.grid() # pravoúhla sieť
ax.get_xticks()

## graf funkcie
ax.plot(X, Y)

zp = newton(f, X, maxiter=50)
zp = np.unique(zp[(zp >= np.amin(X)) & (zp <= np.amax(X))])
for xx in zp:
    ax.plot(xx, 0, 'o', c='red', markersize=6)

### archivácia obrázka
#fig.savefig("<meno súboru>.png")

### samotné zobrazenie
fig.show()
```
