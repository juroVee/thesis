```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
import ipywidgets as widgets
from maux import *
from jedit import editor

#####
##### nakreslenie grafu funkcie
#####

#### vstupnÃ© Ãºdaje
def f(X): return np.sin(2*X) + np.sin(4*X) # ufunc verzia funkcie
X = np.linspace(-np.pi, 2*np.pi, 6*42*100+1) # vÃ½ber hodnÃ´t nezÃ¡vislej premennej pre zaujÃ­mavÃº ÄasÅ¥ grafu
Y = f(X) # odpovedajÃºce hodnoty zÃ¡vislej premennej

def d(X): return 2 * (np.cos(2*X) + 2 * np.cos(4*X))

#### obrÃ¡zok s jednÃ½m diagramom
fig, ax = plt.subplots()
fig.set_size_inches(9, 6) # veÄ¾kosÅ¥ obrÃ¡zka (vÃ½chodzia hodnota je 6x4)

### diagram
init_subplot(ax) # inicializÃ¡cia diagramu: vytvorÃ­ sa pravoÃºhla sÃºradnicovÃ¡ sÃºstava
#ax.set_aspect('equal') # nastavenie rovnakej mierky pre obe osi
ax.set_aspect(1/2*np.pi) # nastavenie rÃ´znej mierky pre obe osi
ax.grid() # pravoÃºhla sieÅ¥

ax.set_title(r"Graf funkcie $y = \sin\ 2x + \sin\ 4x$") # pomenovanie diagramu

## x-ovÃ¡ os
xtick_numerators = range(-2, 4+1)
ax.set_xticks([n * np.pi / 2  for n in xtick_numerators]) # kÃ³tovanie x-ovej osi
ax.set_xticklabels([smart_ticklabel(n, r"\pi", 2)  for n in xtick_numerators]) # oznaÄenie kÃ³t na x-ovej osi

## y-ovÃ¡ os
ax.set_ylim(-2.2, 2.2) # ohraniÄenie hodnÃ´t pre osu y
ax.set_yticks([-2, -1, 0, 1, 2]) # kÃ³tovanie y-ovej osi
## graf funkcie
ax.plot(X, Y)

### archivÃ¡cia obrÃ¡zka
#fig.savefig("<meno sÃºboru>.png")

### samotnÃ© zobrazenie
fig.show()
```
