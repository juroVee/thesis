```python
import numpy as np
%matplotlib notebook
import matplotlib.pyplot as plt
from maux import *
from jedit import editor

#####
##### nakreslenie grafu funkcie
#####

#### vstupnÃ© Ãºdaje
def f(X): return 20 * np.sin(X + 3) * np.cos((X**2) / 2) # ufunc verzia funkcie
X = np.linspace(-5, 5, 10+1) # vÃ½ber hodnÃ´t nezÃ¡vislej premennej pre zaujÃ­mavÃº ÄasÅ¥ grafu
Y = f(X) # odpovedajÃºce hodnoty zÃ¡vislej premennej

#### obrÃ¡zok s jednÃ½m diagramom
fig, ax = plt.subplots()
fig.set_size_inches(7, 9) # veÄ¾kosÅ¥ obrÃ¡zka (vÃ½chodzia hodnota je 6x4)

### diagram
init_subplot(ax) # inicializÃ¡cia diagramu: vytvorÃ­ sa pravoÃºhla sÃºradnicovÃ¡ sÃºstava
ax.set_title(r"Graf funkcie $y = 20 \: sin(x+3) \: cos(\frac{x^2}{2})$") # pomenovanie diagramu
ax.set_aspect('auto') # nastavenie rovnakej mierky pre obe osi
ax.grid() # pravoÃºhla sieÅ¥
ax.get_xticks()

## graf funkcie
ax.plot(X, Y)

### archivÃ¡cia obrÃ¡zka
#fig.savefig("<meno sÃºboru>.png")

### samotnÃ© zobrazenie
fig.show()
```
